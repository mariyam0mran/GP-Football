```python
# Define names for the DataFrames for clearer column suffixes
file_names = [
    "Standard", "Pass", "Time", "Poss", "Shoot", 
    "Misc", "GK", "GKAdv", "GCA", "Def", "PassType"
]

MERGE_KEYS = ['Player', 'Team', 'Season_End_Year']

# Columns that appear in almost every file that we ONLY want from the Standard file
METADATA_COLS = ['Nation', 'Pos', 'Age', 'Born', 'Matches']

def clean_key(val):
    """Removes accents and standardizes strings."""
    if pd.isna(val): return ""
    val = unidecode(str(val))
    return val.strip().lower()
```

```python
def clean_and_prepare_df(df, suffix_name, is_primary_file=False):
    df = df.copy()
    
    # 1. Clean Keys 
    if 'Squad' in df.columns:
        df.rename(columns={'Squad': 'Team'}, inplace=True)
    if 'Team_or_Opponent' in df.columns:
        df = df[df['Team_or_Opponent'].str.lower() == 'team'].drop(columns=['Team_or_Opponent'])
    
    for col in ['Player', 'Team']:
        if col in df.columns:
            df[col] = df[col].apply(clean_key)
            
    df.columns = df.columns.str.strip().str.replace(r'[\(\)/-]', '', regex=True)

 
        
    # --- 2. AGGREGATION ---
    numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
    agg_cols = [c for c in numeric_cols if c not in MERGE_KEYS]
    
    rate_identifiers = ['%', '90', 'per', 'rate', 'avg']
    rate_cols = [c for c in agg_cols if any(x in c.lower() for x in rate_identifiers)]
    count_cols = [c for c in agg_cols if c not in rate_cols]
    
    agg_dict = {}
    for c in count_cols: agg_dict[c] = 'sum'
    for c in rate_cols:  agg_dict[c] = 'mean'

    # Categorical Columns (This now includes Pos, Nation, etc.)
    cat_cols = [c for c in df.columns if c not in agg_cols and c not in MERGE_KEYS]
    for c in cat_cols:
        agg_dict[c] = 'first'
        
    if not df.empty:
        df_unique = df.groupby(MERGE_KEYS).agg(agg_dict).reset_index()
    else:
        df_unique = df

    # 4. Suffix Columns
    rename_dict = {}
    for col in df_unique.columns:
        if col not in MERGE_KEYS: 
            if is_primary_file and col in METADATA_COLS:
                continue 
            rename_dict[col] = f"{col}_{suffix_name}"
            
    return df_unique.rename(columns=rename_dict)
```

```python
# --- Execution ---

processed_dfs = []

for i, df in enumerate(player_files):
    # The first file (Standard) is our "Primary" file
    is_primary = (i == 0)
    suffix = file_names[i]
    
    cleaned_df = clean_and_prepare_df(df, suffix, is_primary_file=is_primary)
    processed_dfs.append(cleaned_df)

# --- Merge ---
player_merged = processed_dfs[0]

for df in processed_dfs[1:]:
    player_merged = pd.merge(
        player_merged,
        df,
        on=MERGE_KEYS,
        how='outer' 
    )

print("✅ Player merge complete:", player_merged.shape)
```

```python
import pandas as pd
import re


# 1. Define columns we WANT to find and consolidate
DESCRIPTIVE_COLUMNS = ['Comp', 'Nation', 'Pos', 'Age', 'Born', 'Url']
MERGE_KEYS = ['Player', 'Team', 'Season_End_Year'] 

# 2. Consolidate Descriptive Columns
# We loop through the list and try to find variations like "Comp_Standard", "Comp_Pass"
for col_name in DESCRIPTIVE_COLUMNS:
    # Match any column that starts with "Comp_" or is exactly "Comp"
    related_cols = [c for c in player_merged.columns if c.startswith(f"{col_name}_")]
    
    if related_cols:
        print(f"Consolidating {col_name} from: {related_cols}")
        
        # Create the consolidated column by taking the first non-null value found
        player_merged[col_name] = player_merged[related_cols].bfill(axis=1).iloc[:, 0]
        
        # Drop the suffixed versions to clean up
        player_merged.drop(columns=related_cols, inplace=True)

# 3. Identify all remaining statistical columns
statistical_cols = [c for c in player_merged.columns if c not in MERGE_KEYS and c not in DESCRIPTIVE_COLUMNS]

# 4. Define the FINAL list of columns to keep
desired_cols = MERGE_KEYS + DESCRIPTIVE_COLUMNS + statistical_cols
final_kept_cols = [c for c in desired_cols if c in player_merged.columns]

# 5. Filter the DataFrame
player_merged = player_merged[final_kept_cols]

# 6. Final Cleanup
# Fill NaNs only in numeric columns
numeric_cols = player_merged.select_dtypes(include=['number']).columns
player_merged[numeric_cols] = player_merged[numeric_cols].fillna(0)

print("✅ Final Column Cleanup Complete:", player_merged.shape)
print("Columns:", player_merged.columns.tolist()[:10], "...")
```

```python
cols = player_merged.columns
seen_cols = {}
new_cols = []

for col in cols:
    # Check if we've seen this column name before
    if col in seen_cols:
        seen_cols[col] += 1
        # Append the count as a suffix
        new_cols.append(f"{col}_{seen_cols[col]}")
    else:
        seen_cols[col] = 0  # Start count at 0
        new_cols.append(col)

# Apply the new unique column list to the DataFrame
player_merged.columns = new_cols
print("Columns successfully made unique.")
```

```python
# Identify and drop the 'Unnamed' columns
cols_to_drop = [c for c in player_merged.columns if 'Unnamed' in c]
player_merged.drop(columns=cols_to_drop, inplace=True, errors='ignore')
```

```python
# --- 2. Suffix Removal ---


# Step A: Identify all "Base Names" (e.g., 'Gls' from 'Gls_Standard')
base_counts = {}
col_mapping = {}

for col in player_merged.columns:
    if col in MERGE_KEYS: continue # Skip keys
    
    # Check if column has a suffix (contains an underscore)
    if '_' in col:
        parts = col.rsplit('_', 1)
        base_name = parts[0]
        
        # We verify if the suffix is one of our known file names to avoid splitting "xG_Expected"
        base_counts[base_name] = base_counts.get(base_name, 0) + 1
        col_mapping[col] = base_name
    else:
        # No underscore? It's already a base name (like "Player")
        base_counts[col] = base_counts.get(col, 0) + 1

# Step B: Rename columns that have a unique Base Name
rename_dict = {}
for original_col, base_name in col_mapping.items():
    if base_counts[base_name] == 1:
        rename_dict[original_col] = base_name

player_merged.rename(columns=rename_dict, inplace=True)
print(f"✅ Renamed {len(rename_dict)} columns to their simple names.")
```

```python
# --- 3. Optimized Manual Consolidation ---

consolidate_cols = ['Mins_Per_90', 'CrdY', 'CrdR'] 

# We will store the new columns and the drop list separately
new_columns_data = {}
all_cols_to_drop = []

for base in consolidate_cols:
    # Find columns like 'CrdY_Standard', 'CrdY_Misc'
    related = [c for c in player_merged.columns if c.startswith(base + '_')]
    
    if len(related) > 0:
        new_columns_data[base] = player_merged[related].bfill(axis=1).iloc[:, 0]
        
        all_cols_to_drop.extend(related)

if new_columns_data:
    # 1. Create a dataframe of the new columns
    new_df = pd.DataFrame(new_columns_data)
    
    # 2. Add them to the main dataframe
    player_merged = pd.concat([player_merged, new_df], axis=1)

# 3. Drop all the old columns at once
player_merged.drop(columns=all_cols_to_drop, inplace=True, errors='ignore')

# 4. Final De-fragmentation (The "Clean Copy" command)
player_merged = player_merged.copy()

print(f"✅ Consolidated {len(new_columns_data)} metrics without fragmentation.")
```

```python
# Clean String Keys
def clean_key(val):
    if pd.isna(val): return ""
    return unidecode(str(val)).strip().lower()


def clean_team_df(df):
    df = df.copy()
    
    # Standardize column names
    if 'Squad' in df.columns:
        df.rename(columns={'Squad': 'Team'}, inplace=True)
    
    if 'Team_or_Opponent' in df.columns:
        # 1. Fill NaNs with 'team' 
        df['Team_or_Opponent'] = df['Team_or_Opponent'].fillna('team')
        
        df = df[df['Team_or_Opponent'].astype(str).str.strip().str.lower() == 'team']
        
        # 3. Drop the column
        df.drop(columns=['Team_or_Opponent'], inplace=True)

    # Apply unidecode to Team names
    df['Team'] = df['Team'].apply(clean_key)
    df['Season_End_Year'] = df['Season_End_Year'].astype(int)
    
    # Clean column headers
    df.columns = df.columns.str.strip()
    
    
    # 1. Metadata columns we want to KEEP
    meta_cols = [c for c in ['Comp', 'Nation'] if c in df.columns]
    keys = ['Team', 'Season_End_Year']
    
    # 2. Check for duplicates
    if df.duplicated(subset=keys).any():
        numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
        non_numeric = df.columns.difference(numeric_cols).tolist()
        
        # Aggregation
        agg_dict = {c: 'mean' for c in numeric_cols if c not in keys}
        for c in non_numeric:
            if c not in keys:
                agg_dict[c] = 'first' # Keep the first label found
                
        df = df.groupby(keys).agg(agg_dict).reset_index()

    
    return df


# Clean all team files
cleaned_files = [clean_team_df(df) for df in team_files]

# Merge

# Start with the first file
team_merged = cleaned_files[0]

# Metadata columns we want to preserve/backfill
meta_cols = ['Comp', 'Nation', 'LgRank']

for i, df in enumerate(cleaned_files[1:]):
    keys = ['Team', 'Season_End_Year']

    
    # Drop shared columns to avoid duplicates, BUT we keep Metadata 
    cols_to_drop = [
        c for c in df.columns 
        if c in team_merged.columns 
        and c not in keys 
        and c not in meta_cols
    ]
    
    if cols_to_drop:
        df = df.drop(columns=cols_to_drop)

    # PERFORM OUTER MERGE
    team_merged = pd.merge(
        team_merged, 
        df,
        on=keys,
        how='outer',
        suffixes=('', '_new') # Keep left clean, mark right as '_new'
    )
    
    # FILL MISSING METADATA
    for col in meta_cols:
        new_col = f"{col}_new"
        if col in team_merged.columns and new_col in team_merged.columns:
            # Fill holes in the main column using the new file's data
            team_merged[col] = team_merged[col].fillna(team_merged[new_col])
            # Drop the temporary '_new' column
            team_merged.drop(columns=[new_col], inplace=True)
            
print("✅ Team merge complete:", team_merged.shape)

# 🧹 4. Final Cleanup
# Fill NaNs in numeric columns
numeric_cols = team_merged.select_dtypes(include=['number']).columns
team_merged[numeric_cols] = team_merged[numeric_cols].fillna(0)

# Sort
team_merged.sort_values(by=['Team', 'Season_End_Year'], inplace=True)

print("✅ Team merge complete:", team_merged.shape)
print("Columns Sample:", team_merged.columns.tolist()[:10])

if 'Comp' in team_merged.columns:
    print("ℹ️ League Column ('Comp') preserved successfully.")
else:
    print("⚠️ Warning: League Column lost!")
```
